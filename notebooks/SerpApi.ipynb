## Funcion export CSV

```python
def export_csv(results, api_call, hospital_folder_name):
    import json
    data = json.loads(json.dumps(dict(results)))
    df = pd.json_normalize(data['reviews'])
    hospital_full_name = hospital_folder_name + "_" + str(api_call)
    hospital_path = f"./{hospital_folder_name}/{hospital_full_name}.csv"
    df.to_csv(hospital_path)
    print(f"Se ha guardado correctamente tu csv en: {hospital_path}")
    
```
